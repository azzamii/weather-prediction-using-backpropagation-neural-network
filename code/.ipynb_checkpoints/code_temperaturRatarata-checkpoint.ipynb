```python
df = df.dropna(axis=0) #drop nilai 0
display(df.info())
display(df)
#df.to_excel('cleansingSuhuRaw.xlsx')
```

```python
df = pd.read_excel('cleansingSuhuFix.xlsx')
display(df.info())
```

```python
x = df.iloc[:,0:1].values
y = df.iloc[:,-1].values
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.75, random_state=0)
```

```python
import tensorflow as tf #bpnn

model = tf.keras.models.Sequential()
bobot = tf.keras.initializers.RandomUniform(minval=0., maxval=1.)
model.add(tf.keras.layers.Dense(units=1))
model.add(tf.keras.layers.Dense(units=6, activation='relu' , kernel_initializer=bobot))
model.add(tf.keras.layers.Dense(units=6, activation='relu', kernel_initializer=bobot))
model.add(tf.keras.layers.Dense(units=1, activation='relu', kernel_initializer=bobot))
model.compile(loss='mean_squared_error', optimizer=tf.keras.optimizers.Adam(0.001))

model.fit(x_train, y_train, epochs=10000, batch_size=100)
```

```python
ypredik = model.predict(x_test)
column_values = [
    'Data fix',
    'Data predict'
]

df = pd.DataFrame(columns = column_values) 
df['Data fix'] = y_test
df['Data predict'] = ypredik
print(df)
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(ypredik, y_test)
print(mse)
```

```python
df['Data fix'].plot(figsize=(16,6), color='blue', label='data aktual')
df['Data predict'].plot(figsize=(16,6), color='red', label='hasil prediksi')
plt.title('hasil prediksi penyinaran suhu')
plt.xlabel('waktu')
plt.ylabel('suhu')
plt.legend(loc='lower left')
```
